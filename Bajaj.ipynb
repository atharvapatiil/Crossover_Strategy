
<h1 align=center><font size=5>Simple Moving Average (SMA) crossover strategy in Python</font></h1>


```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np
```

<h4>Importing Data</h4>

```python
file_path='F:\Python\quandll\csvFile.csv'
df=pd.read_csv(file_path, index_col='Date', parse_dates=True)
df.head()
```

<h4>Plotting Closing Prices Using Seaborn</h4>

```python
plt.figure(figsize=(15,8))
sns.set_style('whitegrid')

sns.lineplot(data=df['Close'])
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.title('Bajaj Electricals Ltd. (500031)',fontsize=20)
plt.xlabel('Period',fontsize=12)
plt.ylabel('Closing Price(INR)',fontsize=12)
```

```python
# Calculating the short-window simple moving average
df['short-window']=np.round(df['Close'].rolling(window=20).mean(), 2)

# Calculating the long-window simple moving average
df['long-window']=np.round(df['Close'].rolling(window=50).mean(), 2)
```

```python
plt.figure(figsize=(15,8))
sns.lineplot(data=df[['Close','short-window','long-window']])
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.title('Bajaj Electricals Ltd. (500031)',fontsize=20)
plt.xlabel('Period',fontsize=12)
plt.ylabel('Price (INR)',fontsize=12)
```
